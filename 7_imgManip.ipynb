```python
# import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt 

# image resizing
img = cv2.imread('cameraman.tif',0)
print(img.shape)
img = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
print(img.shape)


# image rotation
m,n = img.shape

# cols-1 and rows-1 are the coordinate limits.
M = cv2.getRotationMatrix2D(((m-1)/2.0,(n-1)/2.0),90,1)
op_img = cv2.warpAffine(img,M,(m,n))

# plot images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(op_img, cmap = 'gray')
plt.axis('off')

plt.show()

# Affine Transformation

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
op_img = cv2.warpAffine(img,M,(m,n))

# plot images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(op_img, cmap = 'gray')
plt.axis('off')

plt.show()
```

```python
#Edge Detection

# OpenCV inbuilt Sobel edge detection
ox = cv2.Sobel(img,-1,1,0)
oy = cv2.Sobel(img,-1,0,1)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(ox,cmap='gray')
plt.axis(False)

plt.subplot(1,2,2)
plt.imshow(oy,cmap='gray')
plt.axis(False)

out = np.abs(ox) + np.abs(oy)

plt.figure()
plt.imshow(out,cmap='gray')
plt.axis(False)
```
