```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 
import os
import pywt
```

```python
img = cv2.imread('cameraman.tif',0)

plt.imshow(img,cmap='gray')
plt.axis(False)
plt.title('Original Image')

coeffs = pywt.wavedec2(img, 'db1')

arr, coeff_slices = pywt.coeffs_to_array(coeffs)

sort_arr = np.sort(np.abs(arr.reshape(-1)))

thresh = 50

ind = np.abs(arr)>thresh

comp_arr = arr*ind

coeffs_from_arr = pywt.array_to_coeffs(comp_arr, coeff_slices, output_format='wavedecn')

op_img = pywt.waverecn(coeffs_from_arr, wavelet='db1')

plt.figure()
plt.imshow(op_img,cmap='gray')
plt.axis(False)
plt.title('Compressed Image obtained after Wavelet Thresholding')

```
