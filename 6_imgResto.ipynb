```python
# import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# reading image
img = cv2.imread('cameraman.tif',0)

m,n = img.shape

# Adding Guassian Noise
gauss_noise=np.zeros((m,n),dtype=np.uint8)
cv2.randn(gauss_noise,10,50)
gauss_noise=(gauss_noise*0.5).astype(np.uint8)
gn_img=cv2.add(img,gauss_noise)

## Image filtering using smoothing filter

kernel = np.ones((3,3),np.float32)/9
smoothImg = cv2.filter2D(gn_img,-1,kernel)

## Gaussian Filtering
gausfilt = cv2.GaussianBlur(gn_img,(5,5),0)

plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
plt.imshow(img,cmap = 'gray')
plt.axis('off')
plt.title('Original Image')

plt.subplot(2,2,2)
plt.imshow(gn_img,cmap='gray')
plt.axis("off")
plt.title("Noisy Image")

plt.subplot(2,2,3)
plt.imshow(smoothImg,cmap = 'gray')
plt.axis('off')
plt.title('Smoothing')

plt.subplot(2,2,4)
plt.imshow(gausfilt,cmap='gray')
plt.axis("off")
plt.title("Gaussian Filtering")
```

```python
## Salt and Pepper Noise and its removal using median filtering
im = img.flatten()
m = im.shape[0]
print(m)

loc = np.random.randint(0,m,500)

im[loc] = 0

loc = np.random.randint(0,m,500)

im[loc] = 255

plt.figure()
img2 = im.reshape((256,256)).astype(np.uint8)
plt.imshow(img2,cmap='gray')
plt.axis("off")
plt.title("Salt and Pepper (Impulse) Noise")

median = cv2.medianBlur(img2,3)
plt.figure()
plt.imshow(median,cmap='gray')
plt.axis("off")
plt.title("Denoised Image")
```
