```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img =  cv2.imread('peppers.png')

m,n,k = img.shape

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis(False)

#convert color image to grayscale
im_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(im_gray,cmap='gray')
plt.axis(False)
```

```python
## inspecting R, G, B, components of images
plt.figure(figsize=(15,30))

img1 = img[:,:,0]

plt.subplot(1,3,1)
plt.imshow(img1,cmap = 'gray')
plt.axis('off')

img1 = img[:,:,1]
plt.subplot(1,3,2)
plt.imshow(img1,cmap = 'gray')
plt.axis('off')

img1 = img[:,:,2]
plt.subplot(1,3,3)
plt.imshow(img1,cmap = 'gray')
plt.axis('off')
```

```python
## Removing B, G, R, components from images
plt.figure(figsize=(15,30))

img1 = img.copy()
img1[:,:,2] = np.zeros((m,n))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.axis('off')

img1 = img.copy()
img1[:,:,1] = np.zeros((m,n))
plt.subplot(1,3,2)
plt.imshow(img1)
plt.axis('off')

img1 = img.copy()
img1[:,:,0] = np.zeros((m,n))
plt.subplot(1,3,3)
plt.imshow(img1)
plt.axis('off')
```
