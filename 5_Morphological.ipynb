```python
# import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# reading image
img = cv2.imread('bwimg.png',0)

plt.imshow(img,cmap = 'gray')
plt.axis('off')
plt.title('Original Image')

# Perform erosion
# Define structuring element
StrEl = np.ones((5,5),np.uint8)

erosion = cv2.erode(img, StrEl, iterations = 1)


# Perform dilation
# using cv2.getStructuringElement() to define structuring element 
StrEl = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))

dilation = cv2.dilate(img, StrEl, iterations = 1)

# plot image and its histogram
plt.figure(figsize=(15,30))

plt.subplot(1,2,1)
plt.imshow(erosion, cmap = 'gray')
plt.axis('off')
plt.title('Morphological Erosion')

plt.subplot(1,2,2)
plt.imshow(dilation,cmap = 'gray')
plt.axis('off')
plt.title('Morphological dilation')

plt.show()
```

```python
# Perform opening and closing 

# using cv2.getStructuringElement() to define structuring element 
StrEl = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))

opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, StrEl)

closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, StrEl)

# plot image and its histogram
plt.figure(figsize=(15,30))

plt.subplot(1,2,1)
plt.imshow(opening, cmap = 'gray')
plt.axis('off')
plt.title('Morphological Erosion')

plt.subplot(1,2,2)
plt.imshow(closing,cmap = 'gray')
plt.axis('off')
plt.title('Morphological dilation')

plt.show()
```
