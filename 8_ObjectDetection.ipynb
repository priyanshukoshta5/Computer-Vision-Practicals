```python
import time
import numpy as np
import cv2
import matplotlib.pyplot as plt 
from matplotlib.patches import Rectangle
```

```python
# Create our body classifier
car_classifier = cv2.CascadeClassifier('cars.xml')
# Initiate video capture for video file

img =  cv2.imread('car.png',0)

print(type(img),img.shape)
#plt.imshow(img,cmap='gray')
   
cars = car_classifier.detectMultiScale(img, 
                                 scaleFactor=1.1, 
                                 minNeighbors=3)
    
# Extract bounding boxes for any bodies identified
#cv2.imshow('Cars', img)
for (x,y,w,h) in cars:
    plt.imshow(img,cmap='gray')
    plt.axis(False)
    currentAxis = plt.gca()
    currentAxis.add_patch(Rectangle((x, y), w, h, facecolor="none", ec='r', lw=2))
```
